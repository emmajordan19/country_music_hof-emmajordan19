```python
import requests
```

```python
URL = 'https://countrymusichalloffame.org/hall-of-fame/members/'

response = requests.get(URL)
```

```python
type(response)
```

```python
response.status_code
```

```python
from bs4 import BeautifulSoup as BS
```

```python
soup = BS(response.text)
```

```python
soup
```

```python
print(soup.prettify())
```

```python
soup.find('h3')
```

```python
type(soup.find('h3'))
```

```python
members = soup.findAll('h3')
print(type(members))
members
```

```python
artists = []

for member in members:
    artists.append(member.text)
    
artists
```

```python
URL = 'https://countrymusichalloffame.org/hall-of-fame/members/page/'
artists = []

for page in range(1,17):
    new_url = URL + str(page)
    
    response = requests.get(new_url)
    soup = BS(response.text)
    
    names = soup.findAll('h3')
    
    for name in names:
        artists.append(name.text)
    
```

```python
artists
```

```python
page = 1 
year = []
while page != 17:
    URL = f"https://countrymusichalloffame.org/hall-of-fame/members/page/{page}"
    response = requests.get(URL)
    html = response.content
    soup = BS(html, 'lxml')
    for p in soup.findAll('p')[3:12]:
        year.append(p.get_text(strip=True)[9:])
    page = page + 1
```

```python
year
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
hof = pd.DataFrame(list(zip(artists, year)),
                  columns = ['Artist', 'Year'])

hof
```

```python
hof.Year.value_counts().sort_index().plot(kind = 'bar',
                                          figsize = (14, 10),
                                          color = ['purple', 'cornflowerblue'])
plt.title('HoF Inductees per Year',
           fontsize = 18,
           fontweight = 'bold')
```

```python
hof.Year.value_counts().sort_index().plot(kind = 'line',
                                          figsize = (14, 10),
                                          color = ['purple'])
plt.title('HoF Inductees per Year',
           fontsize = 18,
           fontweight = 'bold')
```
